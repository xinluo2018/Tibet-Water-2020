```python
import os
os.chdir('..')
from notebooks import config
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils.metrics_plot import metrics_plot
```

```python
model_name = ['model_single_in_001', 'model_scales_in_001', 'model_scales_gate_001']
metric_name = 'val_loss'
path_metrics_1 = config.root + '/model/pretrained/' + model_name[0] + '_metrics.csv'
path_metrics_2 = config.root + '/model/pretrained/' + model_name[1] + '_metrics.csv'
path_metrics_3 = config.root + '/model/pretrained/' + model_name[2] + '_metrics.csv'

## read metric file
metrics_1 = pd.read_csv(path_metrics_1)
print('metrics_name:',list(metrics_1.columns))
metrics_2 = pd.read_csv(path_metrics_2)
metrics_3 = pd.read_csv(path_metrics_3)
metrics_sel = pd.DataFrame({metric_name+'_'+model_name[0]: metrics_1[metric_name], \
                            metric_name+'_'+model_name[1]: metrics_2[metric_name], \
                            metric_name+'_'+model_name[2]: metrics_3[metric_name]})
metrics_sel.tail()
```

```python
metrics_plot(metrics=metrics_sel, keys=[metric_name+'_'+model_name[0], \
                        metric_name+'_' + model_name[1], metric_name + '_' + model_name[2]], \
             colors=['g', 'r', 'b'],  \
             axis_name=['Iteration', 'Accuracy'],\
             axis_range=[[-5,205], [0., 0.55]], \
             legend_name=[model_name[0], model_name[1], model_name[2]], \
             legend_pos = [0.36, 0.27],\
             smooth_window =21)
```
